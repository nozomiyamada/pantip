```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm
import json, os, re
import numpy as np

from pantip_scrape import get_one_post
```

## instantiate webdriver

```python
driver = webdriver.Chrome()
driver.set_page_load_timeout(15)
```

```python
url = 'https://pantip.com/topic/30000000'
driver.get(url)
```

## function to get post content

```python
soup = BeautifulSoup(driver.page_source)
result = get_one_post(soup) ## import from `pantip_scrape.py`
result
```

## random scrape by ID

```python
file_number = 11
filename = f'data/pantip_random{str(file_number).zfill(2)}.json'

## open scraped file
if os.path.exists(filename):
    with open(filename, 'r', encoding='utf8') as f:
        data = json.load(f)
    i = int(data[-1]['url'].split('/')[-1]) + 50
else:
    with open(f'data/pantip_random{str(file_number-1).zfill(2)}.json', 'r', encoding='utf8') as f:
        data = json.load(f)
    i = int(data[-1]['url'].split('/')[-1]) + 50
    data = []

print('START INDEX :', i)

## scrape
for n in tqdm(range(1001)):
    try:
        url = f'https://pantip.com/topic/{i}'
        driver.get(url)
        soup = BeautifulSoup(driver.page_source)
        result = get_one_post(soup)
        data.append(result)
    except:
        pass
    i += np.random.randint(50, 150)
    
    ## save to json file
    if n % 50 == 0:
        with open(filename, 'w', encoding='utf8') as f:
            json.dump(data, f, indent=2, ensure_ascii=False)
```
